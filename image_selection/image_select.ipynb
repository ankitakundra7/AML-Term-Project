```python
import os
import json
import jsonlines
```

```python
#removing images
#do not run more than once, if need to, be sure to replace the image folder after running this code with the original image folder
location = '/content/drive/My Drive/place directory to img folder here'
count = 0 
delete = 0

for filename in os.listdir(location):
    path = os.path.join(location, filename)  
    if count < delete: 
      os.remove(path)
      print(filename,"has been removed.")
      count += 1 
    else:
      break
```

```python
#counting the number of images 
path, dirs, files = next(os.walk(location))
file_count = len(files)
print(file_count)
```

```python
#getting the image file id in a list
img_list = []
for filename in os.listdir(location):
  img_list.append(filename[:-4])
```

```python
#function accessing and modifying the jsonl files  
def mod_files(file_loc, file_n,out_file):
  path = os.path.join(file_loc, file_n)
  with jsonlines.open(path) as f_in:
    with jsonlines.open(out_file,'w') as f_out:
      for line in f_in:
        if line['id'] in img_list:
          f_out.write(line)
```

```python
loc = '/content/drive/My Drive/place directory to jsonl files here'

mod_files(loc,'train.jsonl','train.jsonl')
mod_files(loc,'test_unseen.jsonl','test_unseen.jsonl')
mod_files(loc,'test_seen.jsonl','test_seen.jsonl')
mod_files(loc,'dev_unseen.jsonl','dev_unseen.jsonl')
mod_files(loc,'dev_seen.jsonl','dev_seen.jsonl')
```
